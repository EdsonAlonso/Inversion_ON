```python
import numpy as np
import matplotlib.pyplot as plt
from random import seed
from auxiliares import generate_L_random_sequences, a_estimator, b_estimator, std_parameter_error
```

```python
N = 50
L = 5
std = 0.3
```

<h1>Generating L pseudo random arrays</h1>

```python
e = generate_L_random_sequences( std, N, L )
```

```python
plt.figure( figsize = (8,8), facecolor = 'w' )
plt.title('Pseudo random sequences', fontsize = 16 )
plt.plot( e.T )
plt.grid( )
plt.show( )
```

<h1>Creating the data and corrupting it with L different noises</h1>

```python
a = 1
b = 3
xobs = np.linspace(0,20,N)
y = a*xobs +  b #dados "verdadeiros" sem ruido
y_noised = e + y
```

<h1>Inverting parameters</h1>

```python
p1 = [ ]
p2 = [ ]
for i in range( L ):
    a_est = a_estimator( xobs, y_noised[ i ] )
    b_est = b_estimator( xobs, y_noised[ i ], a_est )
    p1.append( a_est )
    p2.append( b_est )
```

```python
print( f'p1: {p1}' )
print( f'p2: {p2}' )
```

<h1>Checking the standar deviation error</h1>

```python
std_parameter_error( np.array( [ p1,p2 ] ) )
```
